#Problem statement :
Today, 1.85 million different apps are available on playstore understanding app trends is crucial. This project's objective is to understand and analyze the data to identify this trends.

```python
# importing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# read csv
df=pd.read_csv('/content/googleplaystore.csv')
```

```python
# head
df.head()
```

```python
# tail
df.tail()
```

```python
df.info()
```

```python
# checking if null values present or not
df.isnull().sum()
```

```python
# Dropping price because of all 0 values
df.drop(columns=['Price'],inplace=True)
df
```

```python
# checking if percentage of any column's null values is less than 10
df.isnull().mean()*100
```

```python
# Removing mean to fill it with other null values
df['Rating'].mean()
```

```python
# converting rating to string to replace values
df['Rating']=df['Rating'].astype(str)
```

```python
df.info()
```

```python
# replacing with null values with mean values
df['Rating']=df['Rating'].str.replace('nan','4.19')
df
```

```python
# Converting ratings back to numeric
df['Rating']=pd.to_numeric(df['Rating'])
df.info()
```

```python
df.isnull().mean()*100
```

```python
# dropping null values from particular columns that are less than 10 percent
df.dropna(subset=['Type', 'Content Rating'], inplace=True)
```

```python
# filling null values with unknown
df['Current Ver'].fillna('Unknown', inplace=True)
df['Android Ver'].fillna('Unknown', inplace=True)
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
# Converting reviews to numeric
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df.info()
```

```python
# converting Installs to numeric
df['Installs']=df['Installs'].astype(str)
df['Installs'] = df['Installs'].str.replace('[+,]', '', regex=True).astype(int)
df['Installs']=df['Installs'].astype(int)
```

```python
df.info()
```

```python
# Replacing m and k with nothing and replacing varies with device with nan
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', ''))/1024
    elif size == 'Varies with device':
        return np.nan
    else:
        return np.nan
```

```python
# applying conversion of size
df['Size'] = df['Size'].apply(convert_size)
df['Size'].fillna(df['Size'].mean(), inplace=True)
```

```python
# Checking duplicates
df.duplicated().sum()
```

```python
# Remove duplicate entries
df.drop_duplicates(inplace=True)
```

```python
# Checking duplicates
df.duplicated().sum()
```

```python
# Converting datetime Last Updated to datetime format
df['Last Updated']=pd.to_datetime(df['Last Updated'], errors='coerce')
df
```

```python
# Extracting year column
df['Year']=df['Last Updated'].dt.year
df
```

```python
# Checking counts of each android version
df['Android Ver'].value_counts()
```

```python
# Categorizing android versions
def android_version(version):
    if version.startswith('4.'):
        return '4 and above'
    elif version.startswith('5.'):
        return '5 and above'
    elif version.startswith('6.'):
        return '6 and above'
    elif version.startswith('7.'):
        return '7 and above'
    elif version.startswith('8.'):
        return '8 and above'
    else:
        return 'Varies with device'
df['Android Ver']=df['Android Ver'].apply(android_version)
```

```python
# Checking android version after categorizing
df['Android Ver'].value_counts()
```

# EDA

# Average number of reviews by category

```python
# Grouping by
avg_reviews=df.groupby('Category')['Reviews'].mean().sort_values(ascending=False).head(10)

# Plot
px.bar(x=avg_reviews.index, y=avg_reviews.values, color=avg_reviews.index,
    title='Top 10 Categories by Average Number of Reviews',
    labels={'x': 'Category', 'y': 'Average Number of Reviews'},
)
```

Documentation:
The above graph helps us to determine social app categories have the most active and engaged user based on their average review counts.
